```python
import pandas as pd
import matplotlib.pyplot as plt

# a) Read test and label files
test_data = pd.read_csv('test.csv')
labels = pd.read_csv('test_label.csv')
```

```python
plt.figure(figsize=(25,50))
for i in range(test_data.shape[1]):
    plt.subplot(test_data.shape[1], 1, i+1)
    plt.plot(test_data.iloc[:, i], label=f'Series {i}')
    if labels.iloc[i, 0] == 1:
        plt.axvspan(0, len(test_data), color='red', alpha=0.3)
    plt.legend()
plt.tight_layout()
plt.show()
```

```python
print(test_data.describe())
test_data.hist(figsize=(12, 8))
plt.show()
```

```python
# d) Find out the variables which are the root cause for the anomaly
# Assuming anomalies are labeled with '1' in the labels file
anomaly_indices = labels[labels == 1].dropna().index
root_cause_variables = test_data.iloc[anomaly_indices].idxmax(axis=1)
print(f'Variables that are potential root causes for the anomalies: {root_cause_variables}')
```
